```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
refactor_commits = pd.read_csv("TruePositiveCommits.csv")
```

```python
refactor_commits.head()
```

```python
print(refactor_commits['Unnamed: 11'].dropna().unique())
print(refactor_commits['Unnamed: 12'].dropna().unique())
print(refactor_commits['Unnamed: 13'].dropna().unique())
```

```python
refactor_commits = refactor_commits.drop(columns=["Unnamed: 11", "Unnamed: 12", "Unnamed: 13"])
```

```python
refactor_commits.head()
```

```python
td = refactor_commits["Tech Debt"].fillna("")

# split, trim, lowercase, drop empties
cats = td.str.split("/").explode().str.strip().str.lower()
cats = cats[cats != ""]  # remove blanks

counts = cats.value_counts()
display(counts)  # occurrences across all rows
```

```python
# Count non-NA values in Categorized Refactoring columns
refactor_cols = ["Categorized Refactoring 1", "Categorized Refactoring 2", 
                 "Categorized Refactoring 3", "Categorized Refactoring 4", 
                 "Categorized Refactoring 5", "Categorized Refactoring 6"]
refactor_commits["Number of Refactorings"] = refactor_commits[refactor_cols].notna().sum(axis=1)

# Count Tech Debt values separated by /
refactor_commits["Number of Tech Debts Addressed"] = refactor_commits["Tech Debt"].fillna("").apply(
    lambda x: len([item for item in x.split("/") if item.strip()]) if x else 0
)
```

```python
refactor_commits.to_csv("TruePositiveCommitsClean.csv", index=False)
```
